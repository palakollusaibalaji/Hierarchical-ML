```python
x=df.drop(["country"],axis=1)
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scale = sc.fit_transform(x)
x_scale
```

```python
sample = pd.DataFrame(x_scale,columns=x.columns)
sample
```

```python
from scipy.cluster.hierarchy import dendrogram
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Country dendogram")
dend = shc.dendrogram(shc.linkage(x_scale,method="ward"))
```

```python
from scipy.cluster.hierarchy import linkage
merge = linkage(x_scale,method="complete",metric="euclidean")
plt.title("complete linkage of data")
dendrogram(merge)
plt.show()
```

```python
from scipy.cluster.hierarchy import cut_tree
cluster_tree = cut_tree(merge,n_clusters=5).reshape(-1,1)
cluster_tree
```

```python
sample["Hierarchical Clustering"] = cluster_tree
sample.head(50)
```

```python
sample["Hierarchical Clustering"].value_counts()
```

```python
from sklearn.decomposition import IncrementalPCA
pca = IncrementalPCA(n_components=9)
pca_x = pca.fit_transform(x_scale)
print(pca_x.shape)
```

```python
x_pca = pd.DataFrame(pca_x,columns=["p1","p2","p3","p4","p5","p6","p7","p8","p9",])
x_pca
```

```python
sns.scatterplot(x="gdpp",y="income",data=sample,hue="gdpp",marker="^")
plt.title("GDP",fontsize=20)
plt.xlabel("GDP")
plt.ylabel("ChildMort")
plt.show()
```
